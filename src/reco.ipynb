```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def energy2d(file):
    Ereco = file["RecoEnu"]
    Etrue = file["TrueEnu"]

    plt.hist2d(Ereco, Etrue, bins=100)
    plt.xlim(0.0,8.0)
    plt.ylim(0.0,8.0)
    plt.xlabel("Reconstructed Energy (GeV)")
    plt.ylabel("True Energy (GeV)")
    plt.colorbar()
    # plt.show()

def energy(file):
    Ereco = file["RecoEnu"]
    Etrue = file["TrueEnu"]

    plt.hist(Etrue, bins=200, color='red',alpha=0.4,label="True Energy")
    plt.hist(Ereco,bins=200, color='blue',alpha=0.5,label="Reco Energy")
    plt.xlabel("Neutrino Energy (GeV)")
    plt.ylabel("arb. units")
    plt.legend()
    plt.xlim(0.0,8.0)
    # plt.show()
```

```python
Ar_genie = pd.read_csv('Ar_selection_genie.csv',sep="\t")
C_genie = pd.read_csv('C_selection_genie.csv',sep="\t")
Ar_nuwro = pd.read_csv('Ar_selection_nuwro.csv',sep="\t")
C_nuwro = pd.read_csv('C_selection_nuwro.csv',sep="\t")
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
energy2d(Ar_genie)
plt.title("Argon GENIE")

plt.subplot(2, 2, 2)
energy2d(C_genie)
plt.title("Carbon GENIE")

plt.subplot(2, 2, 3)
energy2d(Ar_nuwro)
plt.title("Argon NuWro")

plt.subplot(2, 2, 4)
energy2d(C_nuwro)
plt.title("Carbon NuWro")

plt.tight_layout()
plt.savefig('2dEnergy.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
energy(Ar_genie)
plt.title("Argon GENIE")

plt.subplot(2, 2, 2)
energy(C_genie)
plt.title("Carbon GENIE")

plt.subplot(2, 2, 3)
energy(Ar_nuwro)
plt.title("Argon NuWro")

plt.subplot(2, 2, 4)
energy(C_nuwro)
plt.title("Carbon NuWro")

plt.tight_layout()
plt.savefig('Energy.png', dpi=600, bbox_inches='tight')
plt.show()
```
